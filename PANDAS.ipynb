# **GROUPBY AND AGGREGATION**

```python
import pandas as pd

data = pd.DataFrame({
    "department": ["IT", "HR", "IT", "HR", "Finance", "IT", "Finance"],
    "employee": ["Raj", "Amit", "Neha", "Priya", "Karan", "Riya", "Vikas"],
    "salary": [60000, 50000, 65000, 52000, 70000, 62000, 72000],
    "age": [25, 30, 28, 32, 35, 26, 38]
})

print(data)
```

```python
dept =data.groupby('department')
dept
for i ,v in dept:
  print(i)
  print(v)
```

```python
dept =data.groupby('department')['salary'].sum()
dept
```

```python
dept =data.groupby('employee')['salary'].sum()
dept
```

```python
dept =data.groupby('department')['age'].sum()
dept
```

```python
dept =data.groupby('department')['age'].mean()
dept
```

```python
data['age'].mean()
```

```python
data['salary'].mean()
```

```python
data['salary'].median()
```

```python
data['salary'].std()
```

```python
data['salary'].agg(['mean','max','min'])
```

# **PIVOT TABLES**

```python
import pandas as pd

data = pd.DataFrame({
    "Date": ["2024-01", "2024-01", "2024-01",
             "2024-02", "2024-02", "2024-02",
             "2024-03", "2024-03"],

    "Region": ["North", "South", "East",
               "North", "South", "East",
               "North", "South"],

    "Product": ["Laptop", "Laptop", "Laptop",
                "Mobile", "Mobile", "Mobile",
                "Tablet", "Tablet"],

    "Sales": [50000, 45000, 40000,
              30000, 28000, 25000,
              20000, 22000],

    "Quantity": [5, 4, 3,
                 10, 9, 8,
                 6, 7]
})

print(data)
```

```python
pd.pivot_table(data,values='Sales',index='Region',columns='Product')
```

```python
pd.pivot_table(data,values='Sales',index='Region',columns='Product',aggfunc='median')
```

```python
pd.pivot_table(data,values=['Sales','Quantity'],index='Region',columns='Product')
```

```python
pd.crosstab(data['Region'],data['Sales'])
```
