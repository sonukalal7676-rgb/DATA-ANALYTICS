# **EXERSISES**

```python
import numpy as np

s = np.array([
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
])

print(s)
```

```python
b = np.sum(s,axis = 1)
b == 45
```

```python
b = np.sum(s,axis = 1)
for i in b:
  if i == 45:
    print("sudoku is valid")
    break
else:
  print("sudoku is invalid")
```

```python
s[0:3,0:3]
```

```python
s[0:3,3:6]
```

```python
s[0:3,6:9]
```

```python
for i in range(0,9,3):
  for j in range(0,9,3):
    n = (s[i:i+3,j:j+3])
    print(n.sum())
```
